```python
section_name = '_whole_time_point'
#########################################################
# loading all the files from file_team to form a file(data_book) in which each file(sheet) is bookkeeping by index 
sheet_name = np.array(['Lesion' + '\n(tissue)', '2wah' + '\n(tissue)', '4wah' + '\n(tissue)', '8wah' + '\n(tissue)',
                       'Control' + '\n(tissue)'])  
######################################################### 
cell_location = '_tissue'
#########################################
total_sheet = len(file_team_order)
data_book = pd.DataFrame()
# set window for loading data
min_index = int(0)
max_index = None #int(-1)
# for bookkeeping the total-index of the data_sheet
data_length = np.array([], dtype = int)
for fn in range(total_sheet):
    # selecting sheet
    tsv_file_raw = pd.read_csv(os.path.join(loading_dir_path, file_team_order[fn]), sep = '\t')[min_index:max_index]
    # selecting 4 columns (nucleotide, amino, cell-number, status)
    tsv_file_in = tsv_file_raw[['nucleotide', 'aminoAcid', 'estimatedNumberGenomes', 'sequenceStatus']]
    # filter out row (keeping 'In' only rows)
    tsv_file = tsv_file_in[tsv_file_in['sequenceStatus'] == 'In'] 
    # selecting 3 columns (nucleotide, amino, cell-number) as a data_sheet
    data_sheet = tsv_file[['nucleotide', 'aminoAcid', 'estimatedNumberGenomes']]
    data_sheet = data_sheet.dropna(axis = 0)
    data_sheet = data_sheet.sort_values(by = 'estimatedNumberGenomes', ascending = False)
    data_sheet = data_sheet.reset_index(drop = True)
    # for bookkeeping the total-index of the data_sheet
    data_length = np.append(data_length, data_sheet.shape[0])
    # linking sheet
    data_book = data_book.append(data_sheet)
# get all non-identical sequences from the raw whole-sequence-set
sequence_all = data_book.drop_duplicates(subset = 'nucleotide')
sequence_all = sequence_all.reset_index(drop = True)
###
# construct all-sequence frame for a sheet (by setting a zero) for non-appearing sequence in the raw frame 
def construct_spectrum_frame(sequence_spectrum, sheet0):
    construct0 = pd.merge(sequence_spectrum, sheet0, on = ['nucleotide', 'aminoAcid'],  how = 'left')
    construct0 = construct0[['nucleotide', 'aminoAcid', 'estimatedNumberGenomes_y']]
    construct0.rename(columns = {'estimatedNumberGenomes_y':'estimatedNumberGenomes', 
                               'time_y':'time'}, inplace = True, copy = False)
    construct0.loc[:,'estimatedNumberGenomes'] = \
    construct0.loc[:,'estimatedNumberGenomes'].apply(lambda x:x if pd.notnull(x) else 0)
    return (construct0)
###
# sheet_0
sheet_i = 0
begin_index = data_length[0:sheet_i].sum()
end_index = data_length[0:sheet_i + 1].sum()
sheetRaw = construct_spectrum_frame(sequence_all, data_book[begin_index : end_index])
# checking the sheet
# setting the display of pd.DataFrame
pd.set_option('display.max_colwidth', 60)
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 30)
pd_show(sheetRaw[0:len(AlvaColorCycle)])
# merge the sheet (pd.merge results in a sorting by the initial sheet_0 because it is the host)
for sheet_i in range(1, total_sheet):
    begin_index = data_length[0:sheet_i].sum()
    end_index = data_length[0:sheet_i + 1].sum()
    sheetRaw_i = construct_spectrum_frame(sequence_all, data_book[begin_index : end_index])
    sheetRaw = pd.merge(sheetRaw, sheetRaw_i, on = ['nucleotide', 'aminoAcid'], how = 'left',
                        suffixes=['', '_{:}'.format(sheet_i)])
# construct array with the associated value of each sequence during all sheets
valueRaw = sheetRaw['estimatedNumberGenomes'].values   
for sheet_i in range(1, total_sheet):
    valueRaw = np.vstack([valueRaw, sheetRaw['estimatedNumberGenomes_{:}'.format(sheet_i)].values])
# get raw data
dataRaw = valueRaw.T
print ('total size of raw-clones = {:}'.format(dataRaw.sum(axis = 0)))
print ('shape of raw-clones = {:}'.format(dataRaw.shape))
###########################################################################################################
#######################################################
### saving size_type_time into table
#######################################################
clone_size = np.zeros(len(file_team_order))
clone_type = np.zeros(len(file_team_order))
for i in range(len(file_team_order)):
    total_raw_clone = dataRaw[:, i]
    total_clone = total_raw_clone[total_raw_clone >= 1]
    clone_size[i] = total_clone.sum() 
    clone_type[i] = total_clone.shape[0]
############################
figure_table = pd.DataFrame({'raw_file': file_team_order,
                              'time (day)': relative_file_day,
                              'sheet_name': sheet_name,
                              'size': clone_size,
                              'type': clone_type})
###
figure_table['person_id'] = figure_table.apply(lambda row: sample_name[sn], axis = 1)
figure_table['location'] = figure_table.apply(lambda row: cell_location, axis = 1)
###
column_name = ['person_id', 'location',
               'raw_file',
               'time (day)',
               'sheet_name',
               'size',
               'type']
### ordering the table by column_name
figure_table = figure_table[column_name]
############################
table_name = '_total_size_type'
file_suffix = '.html'
save_html = os.path.join(saving_dir_path, file_name + section_name + table_name + file_suffix)
figure_table.to_html(save_html)
#table_name = '-total-size-type'
file_suffix = '.csv'
save_csv = os.path.join(saving_dir_path, file_name + section_name + table_name + file_suffix)
figure_table.to_csv(save_csv, index = False)
############################
rrr = pd.read_csv(save_csv)
### setting display
pd.set_option('display.max_colwidth', 100)
pd.set_option('display.max_columns', 10)
pd.set_option('display.max_rows', 6)
pd_show(rrr)
############################
```

```python
section_name = '_circle_packing_demo_1'
# set pCircle
pCircle = []
asteroid_pCircle = []
####################################################################
time_point_title = np.array(['Lesion', '2wah'])
picking_time_point = np.array([0, 1], dtype = int)

total_time_point = len(picking_time_point)
galaxy_diameter = []
total_galaxy_object = []
total_galaxy_asteroid = []
total_galaxy_moon = []
total_galaxy_planet = []
all_asteroid_size = [2, 2, 0]
all_moon_size = [30, 3, 1]
for tn in range(total_time_point):
    data_area = np.copy(dataRaw[:, picking_time_point[tn]]) 
    aPack = alva.AlvaCirclePacking(data_area
                                   , asteroid_size = all_asteroid_size[tn]
                                   , moon_size = all_moon_size[tn])
    total_galaxy_object.append(aPack.total_galaxy_object)
    total_galaxy_asteroid.append(aPack.total_galaxy_asteroid)
    total_galaxy_planet.append(aPack.total_galaxy_planet)
    total_galaxy_moon.append(aPack.total_galaxy_moon)
    galaxy_diameter.append(aPack.galaxy_diameter)
    aPack.out_fit_core()
    # improving by out-core 
    for core in range(aPack.total_coreNum):  
        aPack.out_core(core)
    for core in range(aPack.total_coreNum):  
        aPack.out_core(core)
    for core in range(aPack.total_coreNum):  
        aPack.out_core(core, far_out = True)
    for core in range(aPack.total_coreNum):  
        aPack.out_core(core)
    # patching circle 
    pCircle.append(aPack.patchingCircle(AlvaColorCycle = AlvaColorCycle))
#########################################################################    
### plotting
figure_name = '_Alva'
file_suffix = '.png'
save_figure = os.path.join(saving_dir_path, file_name + section_name + cell_location + figure_name + file_suffix)
AlvaFontSize = 23
numberingFig = numberingFig + 1
wX = total_time_point
wY = 1
figure = plt.figure(numberingFig, figsize = (wX*9, wY*9))
window_title = time_point_title[picking_time_point]
for wn in range(wX*wY):
    window = figure.add_subplot(wY, wX, wn + 1)
    window.add_collection(pCircle[wn])
    window.set_title('$ {:} {:} \ at \ ({:}) $'.format(sample_name[sn], cell_location.replace('_', '\_')
                                                          , window_title[wn]), fontsize = AlvaFontSize)    
    window.set_xlim(-galaxy_diameter[wn]*0.55, galaxy_diameter[wn]*0.55)
    window.set_ylim(-galaxy_diameter[wn]*0.55, galaxy_diameter[wn]*0.55)
    window.set_xticklabels([])
    window.set_yticklabels([])
    window.text(-galaxy_diameter[wn]*0.99*0.55, -galaxy_diameter[wn]*0.99*0.55
                , 'Total Clones = {:}'.format(total_galaxy_object[wn]), fontsize = AlvaFontSize*1)
###
figure.tight_layout()
plt.savefig(save_figure, dpi = 300, bbox_inches = 'tight')
plt.show()
##########
```
